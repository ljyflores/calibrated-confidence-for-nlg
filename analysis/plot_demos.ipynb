```python
bars4 = (
    [0.105, 0.09, 0.08, 0.07, 0.055]
    + [0.5 * i / sum(range(25, 0, -1)) for i in range(25, 0, -1)]
    + [0.1 / 70] * 70
)
barlist = plt.bar(list(range(1, len(bars4) + 1)), bars4, width=0.1)
for i in range(5):
    barlist[i].set_color("b")
for i in range(5, 30):
    barlist[i].set_color("g")
for i in range(30, 100):
    barlist[i].set_color("orange")
plt.gca().set_xticklabels([])

labels = [
    "Good Quality Sequences",
    "Mediocre Quality Sequences",
    "Poor Quality/Ungrammatical Sequences",
]
colors = [
    plt.Rectangle((0, 0), 1, 1, color=label) for label in ["blue", "green", "orange"]
]
plt.legend(colors, labels)

plt.ylabel("Sequence Probability")
plt.xlabel("Sequence")
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(6, 6))

plt.subplot(2, 2, 1)
plt.bar(x=range(10), height=[0.1] * 10)
plt.gca().set_xticklabels([])
plt.ylim(0, 1.1)
plt.xlabel("Class")
plt.ylabel("Class Probability")
plt.title(f"Tail Index: {round(sum([0.1**2] * 10), 2)}")

plt.subplot(2, 2, 3)
values = [np.exp(i) for i in range(11, 1, -1)]
values = [x / sum(values) for x in values]
plt.bar(x=range(10), height=values)
plt.gca().set_xticklabels([])
plt.ylim(0, 1.1)
plt.xlabel("Class")
plt.ylabel("Class Probability")
plt.title(f"Tail Index: {round(sum([x ** 2 for x in values]), 2)}")

plt.subplot(2, 2, 2)
plt.bar(x=range(10), height=[1] + [0.0] * 9)
plt.gca().set_xticklabels([])
plt.ylim(0, 1.1)
plt.xlabel("Class")
plt.ylabel("Class Probability")
plt.title(f"Tail Index: {round(1, 2)}")

plt.subplot(2, 2, 4)
values = [np.exp(i) for i in [8, 8, 7, 6, 5, 4, 3, 2, 1, 0]]
values = [x / sum(values) for x in values]
plt.bar(x=range(10), height=values)
plt.gca().set_xticklabels([])
plt.ylim(0, 1.1)
plt.xlabel("Class")
plt.ylabel("Class Probability")
plt.title(f"Tail Index: {round(sum([x ** 2 for x in values]), 2)}")

plt.tight_layout()
plt.show()
```
